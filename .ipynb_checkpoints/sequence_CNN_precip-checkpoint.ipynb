```python
import matplotlib.pyplot as plt
import numpy as np
from read_data import get_data
import sys
from tqdm import tqdm
from tensorflow.keras.losses import mse
from tensorflow.keras.callbacks import LearningRateScheduler,Callback,EarlyStopping
from tensorflow.keras import Model
import xarray as xr
import cartopy.crs as ccrs


import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import regularizers
from tensorflow.keras import initializers
from tensorflow.keras import backend as K
from sklearn.metrics import r2_score
from netCDF4 import Dataset
# from con1D_model import *
from tensorflow.keras.layers import Dense, Flatten, Layer,Conv1D, Input,Lambda,LeakyReLU,Reshape,Dropout,Concatenate
from tensorflow.keras.initializers import Constant
```

```python
inputs = np.stack(get_data('SAM',200,['qvi','ts','huss','tas'],['qvi','ts','huss','tas'],
                           scaling_method = 'minstd'))

outputs = np.stack(get_data('SAM',200,['pracc'],['pracc'],scaling_method = 'none'))

outputs = np.moveaxis(outputs,0,3)
inputs  = np.moveaxis(inputs,0,3)
```

```python
outputs = outputs[1:,:,:,:]-outputs[:-1,:,:,:]
inputs = inputs[1:,:,:,:]
```

```python
mask =  np.random.rand(inputs.shape[0]) < 0.80
input_train,input_test = inputs[mask,:,:,:], inputs[~mask,:,:,:]
output_train,output_test = outputs[mask,:,:,:], outputs[~mask,:,:,:]
```

```python
def chunk(datain,window):
    nt,nlat,nlon,nfeature = datain.shape
    nchunk = int(nt/window)
    dtin = datain[:nchunk*window ,:,:,:]
    dtin = np.reshape(dtin,(nchunk,window,nlat,nlon,nfeature))
    return dtin
def reshape_data(data):
    nt,window,nlat,nlon,nfeature = data.shape
    
    dtin = np.moveaxis(data,1,3)
    dtin = np.reshape(dtin,(nlat*nlon*nt,window,nfeature))
    return dtin

timeframe=3
x_test = reshape_data(chunk(input_test,timeframe))
x_train = reshape_data(chunk(input_train,timeframe))
y_train = reshape_data(chunk(output_train,timeframe))
y_test = reshape_data(chunk(output_test,timeframe))

```

```python
index = np.where(y_test[:,2,0]>.05)[0]
y_test = y_test[index,:]
x_test = x_test[index,:]
```

```python
units = [200,128,128,64]
l1 = 0e-7

initializer = tf.keras.initializers.GlorotNormal()
batch_size = 128
epochs = 30
activation = tf.nn.relu
output_shape = 1
inshapeX1 = x_train[0,-1,:].shape
inshapeX2 = x_train[0,:,:1].shape

val_split = 0.2


input_layer = Input(shape=(inshapeX1), name='inputs')
input_cnn = Input(shape=(inshapeX2), name='incnn')
conv = Conv1D(filters = 3, kernel_size=3,name = 'conv')(input_cnn)
xc = Flatten()(conv)
xs = Concatenate(axis = -1)([xc,input_layer])
xs =  Dense(units=units[0],activation = activation,kernel_initializer=initializer,
            kernel_regularizer=regularizers.l1(l1=l1),name = 'dense1')(xs)

xs =  Dense(units=units[1],activation = activation,kernel_initializer=initializer,
            kernel_regularizer=regularizers.l1(l1=l1),name = 'dense2')(xs)

xs =  Dense(units=units[2],activation = activation,kernel_initializer=initializer,
            kernel_regularizer=regularizers.l1(l1=l1),name = 'dense3')(xs)

xs =  Dense(units=units[3],activation = activation,kernel_initializer=initializer,
            kernel_regularizer=regularizers.l1(l1=l1),name = 'dense4')(xs)

y_predict =  Dense(units=output_shape,activation = None,name = 'prediction')(xs)

print (y_predict.shape)


nn_precip = Model(inputs = [input_layer,input_cnn],
              outputs = [y_predict]) 

opt = keras.optimizers.Adam(learning_rate=0.0001)

nn_precip.compile(optimizer = opt,loss = 'mse')

earlyStopping=keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, verbose=0, mode='auto')



hist = nn_precip.fit({'inputs' : x_train[:,-1,:],'incnn':x_train[:,:,:1]},
                  {'prediction': y_train[:,2,0]},
                   epochs = epochs,  validation_split = val_split,callbacks= [earlyStopping],
                     batch_size = batch_size,verbose = 1, shuffle=True)
```

```python
y_pred = nn_precip.predict([x_test[:,-1,:],x_test[:,:,:1]])
```
