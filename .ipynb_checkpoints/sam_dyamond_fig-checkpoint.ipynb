```python
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
from typhon.cloudmask import scai
from tqdm import tqdm
import os
os.environ['PROJ_LIB'] = '/home/sara/anaconda3/share/proj'
from mpl_toolkits.basemap import Basemap

import matplotlib.pyplot as plt
```

```python
ds_cl = Dataset('/media/sara/LaCie/DYAMOND/cloud_condensate_ice.nc')
ds_cll = Dataset('/media/sara/LaCie/DYAMOND/cloud_condensate.nc')
ds_cl = Dataset('ww68.nc')
ds_cll = Dataset('cc68.nc')
```

```python
ds_olr =  Dataset('/media/sara/LaCie/DYAMOND/olr.nc')
```

```python
cl = ds_cl.variables['clivi'][0,:,:]
cll = ds_cll.variables['clcvi'][0,:,:]
```

```python
lat = ds_cl.variables['lat'][:]
lon = ds_cl.variables['lon'][:]
```

```python
cl_mask = np.copy(cll)
cl_mask+=cl
mask2 = np.copy(cl_mask)
cl_mask[(cll + cl)<=0.71] = np.nan
mask2[(cll + cl)<=0.31] = np.nan
mask2[(cll + cl)>=0.71] = np.nan
```

```python
fig = plt.subplots(nrows = 1 , ncols = 1,figsize= (17,10), )
# setup Lambert Conformal basemap.
# set resolution=None to skip processing of boundary datasets.
m = Basemap(projection='cyl',
            llcrnrlat=-60, urcrnrlat=60,
            llcrnrlon=0, urcrnrlon=360)
x,y = np.meshgrid(lon,lat)
xx, yy = m(x,y)
vect  = np.arange(0,1.01,0.1)
th1 = 0.1
dth = 0.1
for i in range(2,7):
    mask = np.zeros_like(cl)
    mask+=1 

    if i>=6:
        mask[(cll + cl)<=i*th1] = np.nan
    else: 
        mask[(cll + cl)<=i*th1] = np.nan
        mask[(cll + cl)>=i*th1+dth] = np.nan
    
    m.contourf(xx,yy,mask,vect,cmap = 'gray',alpha = (2*i-2)*th1)

m.bluemarble()
plt.text(0,-66,'SAM DYAMOND 4km Resolution ',color = 'black',fontsize= 16,fontname = 'Helvetica')
plt.text(0,-73,'24-02-2020 at 10:00 ',color = 'black',fontsize= 16,fontname = 'Helvetica')

plt.show()
```
