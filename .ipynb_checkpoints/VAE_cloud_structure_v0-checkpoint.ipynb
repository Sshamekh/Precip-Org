SAM data: degree in longitude is constant as the radius changes. This means moving poleward, the models resolution increases. The latitudinal resolution has been reduced moving poleward in order for the models grid to stay close to square shape. Models resolution near the equator is about 4.3 and it reduces to 4.1 at 20 S/N. 

Now, to coarse grain the data, we approxmate models resolution to be 4.2 km in both lon lat direction. 
Then we decide the rescalaing size. This is our large scale, or GCM size grid. Here I represent it with largescale. 

```python
# %%time
import sys, importlib
importlib.reload(sys.modules['utils_vae'])
from utils_vae import train_test_data
t1 = 0
t2 = 1
high_res = 4.0 # km 
large_scale = 100 #km
x_hr,x_lg,y_out = train_test_data(path, high_res,large_scale,t1,t2,threshold_precip = 0.1)
```

```python
mask = np.random.rand(x_hr.shape[0])<0.8
x_train_hr,x_test_hr = x_hr[mask,:],x_hr[~mask,:]
x_train_lg,x_test_lg = x_lg[mask,:],x_lg[~mask,:]
y_train_lg,y_test_lg = y_out[mask,:],y_out[~mask,:]
```

```python
# # Opening JSON file
# with open('model_config05.json') as json_file:
#     model_config = json.load(json_file)
model_config = {'inshape': [4],
 'configs_encoder': {'activation': 'relu',
  'latent_dim': 4,
  'conv_1': {'filter_num': 16, 'kernel_size': [4, 4], 'stride': 2},
  'conv_2': {'filter_num': 16, 'kernel_size': [4, 4], 'stride': 2},
  'conv_3': {'filter_num': 8, 'kernel_size': [4, 4], 'stride': 2},
  'dense_1': {'unit_num': 128},
  'dense_2': {'unit_num': 64},
  'dense_3': {'unit_num': 32},
  'dense_4': {'unit_num': 16},
  'dense_5': {'activation': None}},
 'config_decoder': {'latent_dim': 4,
  'activation': 'relu',
  'dense1': {'unit_num': 64},
  'dense2': {'unit_num': 128},
  'dense3': {'unit_num': 265},
  'dense4': {'unit_num': 512},
  'dense5': {'unit_num': 625}},
 'config_dense': {'activation': 'relu',
  'layer_1': {'unit_num': 128},
  'layer_2': {'unit_num': 128},
  'layer_3': {'unit_num': 128},
  'layer_4': {'unit_num': 64},
  'layer_5': {'unit_num': 1, 'activation': None}},
 'config_cloudnn': {'activation': 'relu',
  'layer_1': {'unit_num': 64},
  'layer_2': {'unit_num': 32},
  'layer_3': {'unit_num': 8},
  'layer_4': {'unit_num': 1, 'activation': None}}}
```

```python
# model_config["configs_encoder"]['latent_dim'] + inshape

def loss_reconstruction_z(latent_var_z, latent_var_z_rotated,  lambda_weight):
    def _zr_reconstruction_loss(true, pred):
        """
        """
        print ("len:" , len(true))
        print ("shape true1:",true[1].shape)
        mse_cloud  =  K.square(true[1] - pred[1])
        mse_precip =  K.square(true[0] - pred[0])
        mse_hr     =  K.sum(K.square(true[2] - pred[2]), axis=(1,2))
        
        print ('cloud mse shape', tf.shape(mse_cloud))
        print ('precip mse shape', tf.shape(mse_precip))
        print ('high res mse shape', tf.shape(mse_hr))
        
        
        
        rotation_constraint = K.sum(K.square(latent_var_z - latent_var_z_rotated)/  K.std(latent_var_z,axis = 0),axis =1)
        
        print ('rotation constraint shape', tf.shape(rotation_constraint))
        
        return K.mean(mse_cloud + mse_precip + mse_hr + lambda_weight * rotation_constraint)
    return _zr_reconstruction_loss
```

```python
coef = 1e1
inshape = 4
inshape_cloud = (inshape+model_config["configs_encoder"]['latent_dim'],)
inshape_precip = (inshape+model_config["configs_encoder"]['latent_dim']+1,)

## rotating data in batch while training 
data_augmentation = tf.keras.Sequential([
  layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
    layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
#   layers.experimental.preprocessing.RandomRotation(1,fill_mode= 'wrap',
#     interpolation='bilinear',
# )
])

encoder_result = encoder_gen((25, 25,1),model_config["configs_encoder"])
decode_zz = decoder_dense((25, 25,1),model_config["config_decoder"])

clouds_cc = cloud_model(inshape_cloud,model_config["config_cloudnn"])
dense_model = dense_gen(inshape_precip,model_config["config_dense"])

inputhighres = tf.keras.layers.Input(shape=(25, 25,1))
input_rotate = data_augmentation(inputhighres)
inputdense = tf.keras.layers.Input(shape=(inshape,))




z_orig = encoder_result.vae_encoder(inputhighres)
z_rotated= encoder_result.vae_encoder(input_rotate)

hr_reconst= decode_zz(z_orig)
# hr_reconst_r= decode_zz(zr)

z_clt = tf.keras.layers.Concatenate(axis = -1)([z_orig,inputdense])
cloud_cover = clouds_cc(z_clt)



zz = tf.keras.layers.Concatenate(axis = -1)([z_orig,inputdense,cloud_cover])
print (zz.shape,z_orig.shape,z_clt.shape)
precip = dense_model.dense_nn(zz)
vae = tf.keras.Model(inputs=[inputhighres,inputdense], outputs=[precip,cloud_cover,hr_reconst])

optimizer = tensorflow.keras.optimizers.Adam(lr=0.0004)
callback_lr=LearningRateScheduler(schedule,verbose=1)
earlyStopping=tensorflow.keras.callbacks.EarlyStopping(monitor='val_loss', patience=7, verbose=0, mode='auto')
vae.compile(loss = loss_reconstruction_z(
            z_orig,
            z_rotated,
            coef), optimizer = optimizer)


# std_z = tf.math.reduce_std(z, axis=0, keepdims=True, name=None)
# RI_loss = K.mean(K.square(z-zr)/std_z)

# print (std_z.shape,(K.square(z-zr)).shape)
# print(RI_loss.shape)
# vae.add_metric(RI_loss,name = 'RI_loss')
# vae.add_loss(coef*RI_loss)
```

```python
%%time
hist = vae.fit(
        x=[x_train_hr,x_train_lg[:,:-1]], 
        y=[y_train_lg[:,1],y_train_lg[:,0],x_train_hr], 
        epochs=100,
        batch_size=128,
        validation_split = 0.3,
        callbacks= [earlyStopping,callback_lr],
        shuffle = True
    )
```
