```python
def make_mask(data,rlow,rhigh):
    mask1 = np.zeros_like(data)
    mask2 = np.zeros_like(data)
    mask1[data>rlow] = 1
    mask2[data<rhigh] = 1 
    mask = mask1*mask2
    return mask
def bin_threshold(data1,data2,binsize):
    minpw = data1.min()
    maxpw = data1.max()
    nbins = int((maxpw - minpw)/binsize)
    bined_pw = []
    bined_precip = []
    for i in range(nbins+1):
        th1 = minpw+i*binsize 
        th2 = minpw+(i+1)*binsize 
        indecies = np.where((data1>=th1) & (data1<th2))[0]
        if len(indecies)>5:
            bined_pw.append(np.mean(data1[indecies]))
            bined_precip.append(np.mean(data2[indecies]))
    return bined_pw,bined_precip
# binpw,binprec = bin_threshold(pwocean[mask!=0],precip_ocean[mask!=0],1)
```

```python
fig, ax = plt.subplots(nrows = 1 , ncols = 1, figsize=(10,8))
for tht in range(263,271):
    mask = make_mask(ta_ocean,tht,tht+1)
    binpw,binprec = bin_threshold(pwocean[mask!=0],4*precip_ocean[mask!=0],0.3)    
    if tht ==269:
        ax.scatter(pwocean[mask!=0],4*precip_ocean[mask!=0],label = 'T '+str(tht),alpha = 0.3)
    ax.plot(binpw,binprec,label = 'T '+str(tht))
    
    
    ax.set_ylim(0,8)
ax.legend(fontsize = 14)
ax.set_xlabel('PW',fontsize = 14)
ax.set_ylabel('Precip',fontsize = 14)
ax.tick_params(axis='both', which='major', labelsize=14)
# ax.set_ylim(0,0.1)
```
