```python
def R2_bins(ds,bins):
    
    data  = ds[:,0]
    datat = ds[:,1]
    datap = ds[:,2]
    
    bb,edge = np.histogram(data,bins = bins)
    r2_return = np.zeros((bins))
    ptrue_bins =[]
    p_predict_bins=[]
    for i in range(bins): 
        
        edge1 = edge[i]
        edge2 = edge[i+1]
        inds = np.where((data>=edge1) & (data< edge2))[0]
        if (len(inds)>50):
            d_1 = datat[inds]
            d_2 = datap[inds]

            r2_return[i] =  r2_score(d_1,d_2)
            
            ptrue_bins.append(d_1)
            p_predict_bins.append(d_2)
        
    return r2_return,edge,bb
        
# pp4_16,bin_pw,bb = R2_bins(ds4_16,100)
# pp4_24 = R2_bins(ds4_24,100)
# pp2_24 = R2_bins(ds2_24,100)
pp4_32,_,_ = R2_bins(ds4_32,100)
pp2_32,_,_ = R2_bins(ds2_32,100)
# pp4_24_2 = R2_bins(ds4_24_2,100)
# pp4_48 = R2_bins(ds4_48,100)
```
